```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import skimage.io
%matplotlib inline
```

```python
#Problem 2a
def masterEq(conc, k, dt):
    numBoxes, timePoints = np.shape(conc)
    for t in range(1, timePoints):
        conc[0, t] = conc[0, t-1] + k * dt * conc[1, t-1] - k * dt * conc[0, t-1] + dt * (2 * 10 ** 4)
        last = numBoxes - 1
        conc[last, t] = conc[last, t-1] + k * dt * conc[last-1, t-1] - k * dt * conc[last, t-1]
        for box in range(1, numBoxes - 1):
            conc[box, t] = conc[box, t-1] + k * dt * conc[box-1, t-1] + k * dt * conc[box+1, t-1] -\
            2 * k * dt * conc[box, t-1] - (dt * conc[box, t - 1] / tau)
    return conc
```

```python
k = .05
dt = 1 / 20
numBoxes = 50
timeSteps = 60001 
tau = 3000
conc = np.zeros((numBoxes, timeSteps)) 
initial_position = 0 
conc[initial_position, 0] = 7.5E5
conc = masterEq(conc, k, dt)
```

```python
time_points = [1 * 60 ,10 * 60, 30 * 60, 100 * 60, 300 * 60, 1000 * 60]
bins = np.arange(0, numBoxes, 1)
fig, ax = plt.subplots(nrows=len(time_points), ncols=1, sharex=True, sharey=True,
                      figsize=(5, 6))

for i in range(len(time_points)):
    ax[i].bar(bins, conc[:, time_points[i]])
    ax[i].set_title('$t$ = %s min' % (time_points[i] / 60))
    ax[i].set_ylabel('$p(x)$')
ax[-1].set_xlabel('box number')
plt.tight_layout()
```

```python
for i in range(0, 6):
    plt.plot(conc[:, time_points[i]], label='time = ' + str(time_points[i] / 60) + ' minutes')
    plt.yscale('log')
plt.legend()
```

```python
#Over the timescale of 1000 minutes, our profile does not reach steady state. Since the cell cycle takes 10 minutes,
#it would take over 100 cycles to reach steady state. The gradient (concentration over x axis) is already pretty stable 
#by the 10th cell cycle, so therefore, steady state will never be reached. 
```

```python
#Problem 3a
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import skimage.io

%matplotlib

embryo1 = skimage.io.imread('desktop/bi1/0.5xBcd.tif')

with sns.axes_style("white"):
    plt.imshow(embryo1)
    
clicksA = plt.ginput(2)
xa1 = clicksA[0][0]
ya1 = clicksA[0][1]
xa2 = clicksA[1][0]
ya2 = clicksA[1][1]

with sns.axes_style("white"):
    plt.plot((xa1, xa2), (ya1, ya2), color='r', marker='o')

clicksB = plt.ginput(2)
xb1 = clicksB[0][0]
yb1 = clicksB[0][1]
xb2 = clicksB[1][0]
yb2 = clicksB[1][1]

with sns.axes_style("white"):
    plt.plot((xb1, xb2), (yb1, yb2), color='b', marker='o')

furrow_click = plt.ginput(1)
furrX = furrow_click[0][0]
furrY = furrow_click[0][1]

embryoLength = np.sqrt((xa2 - xa1) ** 2 + (ya2 - ya1) ** 2)
furrAnt = np.sqrt((furrX - xa1) ** 2 + (furrY - ya1) ** 2)

relFurrPos1 = furrAnt / embryoLength

print('Relative furrow position is: ' + str(relFurrPos1))
```
