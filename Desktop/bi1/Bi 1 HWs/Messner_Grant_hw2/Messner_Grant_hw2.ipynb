```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Problem 1g
dt = .1
totalTime = 40
time_range = np.arange(0, totalTime, dt)
```

```python
n_t=np.zeros(len(time_range))
```

```python
N0 = 2
n_t[0]=N0
```

```python
k = .5
for i in range(len(time_range) - 1):
    dN = k * n_t[i] * dt 
    n_t[i + 1] = n_t[i] + dN
```

```python
plt.plot(time_range, n_t, '-', label='Sea otters')
plt.xlabel('time (years)')
plt.ylabel('Number of otters')
plt.legend()
```

```python
# Problem 1h
n2_t=np.zeros(len(time_range))
n2_t[0] = 2
for i in range(len(time_range) - 1):
    dN = k * n2_t[i] *(1 - n2_t[i]/60000) * dt 
    n2_t[i + 1] = n2_t[i] + dN
```

```python
plt.plot(time_range, n2_t, '-', label='Sea otters')
plt.xlabel('time (years)')
plt.ylabel('Number of otters')
plt.legend()
```

```python
#Problem 1i, 5 whales
n3_t=np.zeros(len(time_range))
n3_t[0] = 60000
k2 = .025 # predation rate constant
for i in range(len(time_range) - 1):
    dN = (k * n3_t[i] *(1 - n3_t[i]/60000 - 5*k2))* dt #5 whales present
    n3_t[i + 1] = n3_t[i] + dN
```

```python
plt.plot(time_range, n3_t, '-', label='Sea otters')
plt.xlabel('time (years)')
plt.ylabel('Number of otters')
plt.legend()
```

```python
#Problem 1i, 10 whales
n4_t=np.zeros(len(time_range))
n4_t[0] = 60000
k2 = .025 # predation rate constant
for i in range(len(time_range) - 1):
    dN = (k * n4_t[i] *(1 - n4_t[i]/60000 - 10*k2))* dt #10 whales now present
    n4_t[i + 1] = n4_t[i] + dN
```

```python
plt.plot(time_range, n4_t, '-', label='Sea otters')
plt.xlabel('time (years)')
plt.ylabel('Number of otters')
plt.legend()
```
