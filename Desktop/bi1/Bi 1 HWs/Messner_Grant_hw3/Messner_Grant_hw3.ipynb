```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dt = .1
totalYears = 60
time_range = np.arange(0, totalYears, dt)
```

```python
male = np.zeros(len(time_range))
female = np.zeros(len(time_range))
male[0] = 10
female[0] = 7
for i in range(0 , len(time_range) - 1):
    if((i / 10) < 15):
        male[i + 1] = male[i]
    else:
        male[i + 1] = male[i] + .45
    female[i + 1] = female[i]
```

```python
plt.plot(time_range, male,'-', label='Sperm mutations')
plt.plot(time_range, female,'-', label='Egg mutations')
plt.xlabel('Time (years)')
plt.ylabel('Number of mutations')
plt.legend()
```
