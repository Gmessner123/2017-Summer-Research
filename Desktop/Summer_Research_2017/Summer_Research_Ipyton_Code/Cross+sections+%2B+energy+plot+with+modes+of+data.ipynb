```python
def gwFit(r, t):
    return (1.50 * 10 **85) * (t) / (((695000 * 10 ** 3) * r) ** 5)
```

```python
T = 10 ** 1
plt.figure(figsize=(15,15))

e = 2.718281828459045
fit = np.polyfit(log_r_min_PN, log_deltaE_PN, 1, cov=True)
fit2 = np.polyfit(log_r_min, log_deltaE, 1, cov=True)
fit3 = np.polyfit(log_r_min_PN_2, log_deltaE_PN_2, 1, cov=True)
cov1 = fit[1]
cov2 = fit2[1]
cov3 = fit3[1]
fitVals_x_PN = []
fitVals_y_PN = []
fitVals_x = []
fitVals_y= []
fitVals_x_PN_2 = []
fitVals_y_PN_2 = []
fitVals_gw_x = []
fitVals_gw_y = []

for i in range(-4, 1):
    fitVals_x_PN.append(e ** i)
    fitVals_y_PN.append(e ** (fit[0][0] * i) * e ** fit[0][1])
    fitVals_x.append(e ** i)
    fitVals_y.append(e ** (fit2[0][0] * i) * e ** fit2[0][1])
    fitVals_x_PN_2.append(e ** i)
    fitVals_y_PN_2.append(e ** (fit3[0][0] * i) * e ** fit3[0][1])
    fitVals_gw_x.append(e ** i)
    fitVals_gw_y.append(gwFit(e ** i, T))

plt.plot(fitVals_x_PN, fitVals_y_PN, label = '\n Relativistic best fit line: \n '
        'ln($\Delta$E) = -1.66 ln($r_{min}$ / $r_{sun}$) + 104.36' 
         '\n $R^{2}$ = .096', color = 'red')
plt.plot(fitVals_x, fitVals_y, label = '\n Nonrelativistic best fit line: \n '
        'ln($\Delta$E) = -0.196 ln($r_{min}$ / $r_{sun}$) + 97.713'
         '\n $R^{2}$ = .014', color = 'blue')
plt.plot(fitVals_x_PN_2, fitVals_y_PN_2, label = '\n Relativistic best fit line (PN2.5 only): \n '
        'ln($\Delta$E) = -3.52 ln($r_{min}$ / $r_{sun}$) + 100.06'
         '\n $R^{2}$ = .992', color = 'green')
plt.plot(fitVals_gw_x, fitVals_gw_y, label = '\n GW energy loss prediction: \n '
       '$\Delta$E = (3.26 * $10^{85}$)t / $r_{min}^{5}$ '
         '\n t = $10^{1}$ sec \n', color = 'orange')
    
plt.plot(r_min_PN, deltaE_PN, 'o', label = 'Relativistic', color = 'red')
plt.plot(r_min, deltaE, 'o', label = 'Nonrelativistic', color = 'blue')
plt.plot(r_min_PN_2, deltaE_PN_2, 'o', label = 'Relativistic PN2.5 only', color = 'green')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$r_{min}$ / $r_{sun}$')
plt.ylabel('Energy loss $\Delta$E (erg)')
plt.legend()

print(fit)
print(fit2)
print(fit3)
print(np.sqrt(np.diag(cov1)))
print(np.sqrt(np.diag(cov2)))
print(np.sqrt(np.diag(cov3)))
```
