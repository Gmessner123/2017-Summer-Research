```python
def gwFit(r, t):
    return (3.26 * 10 **76) * (t) / (((695000 * 10 ** 3) * r) ** 5)
```

```python
T = 10 ** -1
plt.figure(figsize=(15,15))
e = 2.718281828459045
plt.plot(small_r_min_PN, DeltaE_smallr_PN, 'o', label = 'Relativistic data', color = 'blue')
plt.plot(small_r_min, DeltaE_smallr, 'o', label = 'Nonrelativistic data', color = 'red')
fit = np.polyfit(log_small_r_min_PN, log_DeltaE_smallr_PN, 1, cov=True)
fit2 = np.polyfit(log_small_r_min, log_DeltaE_smallr, 1, cov=True)
cov1 = fit[1]
cov2 = fit2[1]
fitVals_x_PN = []
fitVals_y_PN = []
fitVals_x = []
fitVals_y= []
fitVals_gw_x = []
fitVals_gw_y = []
for i in range(- 8, -3):
    fitVals_x_PN.append(e ** i)
    fitVals_y_PN.append(e ** (fit[0][0] * i) * e ** fit[0][1])
    fitVals_x.append(e ** i)
    fitVals_y.append(e ** (fit2[0][0] * i) * e ** fit2[0][1])
    fitVals_gw_x.append(e ** i)
    fitVals_gw_y.append(gwFit(e ** i, T))
    
#t1 = np.arange(e ** (-8.0), e ** (-3.0), 0.1)
#plt.plot(t1, gwFit(t1, 10^-3), 'k')

plt.plot(fitVals_x_PN, fitVals_y_PN, label = '\n Relativistic best fit line: \n '
        'ln($\Delta$E) = -3.626 ln($r_{min}$ / $r_{sun}$) + 81.237'
         '\n $R^{2}$ = .975', color = 'blue')
plt.plot(fitVals_x, fitVals_y, label = '\n Nonrelativistic best fit line: \n '
        'ln($\Delta$E) = 0.192 ln($r_{min}$ / $r_{sun}$) + 96.715'
         '\n $R^{2}$ = .034', color = 'red')

plt.plot(fitVals_gw_x, fitVals_gw_y, label = '\n GW energy loss prediction: \n '
       '$\Delta$E = (3.26 * $10^{76}$)t / $r_{min}^{5}$ '
         '\n t = $10^{-1}$ sec', color = 'green')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('$r_{min}$ / $r_{sun}$')
plt.ylabel('Energy loss $\Delta$E (erg)')
plt.legend()
print(fit)
print(fit2)
print(np.sqrt(np.diag(cov1)))
print(np.sqrt(np.diag(cov2)))
#Note: After linear regression t-test it was found that the nonzero slope of the nonrelativistic energy curve was
#statistically insignificant at any reasonable significance level (p=.253). The relativistic curve was statistically
#significant at any reasonable alpha level (p~10^-13). 
```
