```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import subprocess
import random
```

```python
def getLenNum(s):
    for i in range(len(s) - 1):
        if(s[i:i+1] == ' ' or s[i:i+1] == '\\'):
            return i
    return(len(s) - 1)
```

```python
n = 10000
file = 'desktop/logfile.txt'
f = open(file, 'r')
b = []
v = []
m = []
m0 = []
m1 = [] 
r = []
r0 = []
r1 = []
a = []
e = []
global line 
line = 0

def readline2():
    global line
    line += 1
    return f.readline()

readline2()
i = 0
while(i < n):
    classification = readline2()
    t = float(classification[10:])
    if(t < .0187):
        l = readline2()
        while(l[0:1] != '*'):
            l = readline2()
        continue
        
    if classification[5:7] == 'BB':
        l = readline2()
        while(l[0:1] != '*'):
            l = readline2()
        i += 1
        continue
        
    s = readline2()
    s2 = readline2()
    s3 = readline2()
    for j in range(len(s) - 1):
        sub = s[j : j + 1]
        if sub == 'b' and s[j+1 : j+2] != 'i':
            stop = getLenNum(s[j+2:])
            b.append(s[j+2 : j + 2 + stop])
        if sub == 'v':
            stop = getLenNum(s[j+2:])
            v.append(s[j+2 : j + 2 + stop])
            break
    
    for j in range(len(s2) - 1):
        sub = s2[j : j + 1]
        if sub == 'm' and s2[j+1 : j+2] != 's':
            stop = getLenNum(s2[j+2:])
            m.append(s2[j+2 : j + 2 + stop])
        if sub == 'R':
            stop = getLenNum(s2[j+2:])
            r.append(s2[j+2 : j + 2 + stop])
            break
            
    for j in range(len(s3) - 2):
        sub2 = s3[j:j+2]
        if sub2 == 'm0':
            stop = getLenNum(s3[j+3:])
            m0.append(s3[j+3: j + 3 + stop])
        if sub2 == 'm1':
            stop = getLenNum(s3[j+3:])
            m1.append(s3[j+3: j + 3 + stop])
        if sub2 == 'R0':
            stop = getLenNum(s3[j+3:])
            r0.append(s3[j+3: j + 3 + stop])
        if sub2 == 'R1':
            stop = getLenNum(s3[j+3:])
            r1.append(s3[j+3: j + 3 + stop])
            break
    
    for j in range(len(s3) - 1):
        sub = s3[j : j + 1]
        if sub == 'a' and s3[j+1 : j+2] == '=':
            stop = getLenNum(s3[j+2:])
            a.append(s3[j+2 : j + 2 + stop])
        if sub == 'e' and s3[j+2 : j+3] != 'b' and s3[j+1 : j+2] == '=':
            stop = getLenNum(s3[j+2:])
            e.append(s3[j+2 : j + 2 + stop])
            break
    
    l = readline2()
    while(l[0:1] != '*'):
        l = readline2()
    i += 1

f.close()

def stringToFloat(li):
    floats = []
    for i in range(len(li)):
        trim = ''
        string = li[i]
        j = 0
        if(string != '0'):
            while j < len(string) - 1 and string[j : j + 1] != '\\' and string[j : j + 1] != 'R'and string[j : j + 1] != 'E':
                trim += string[j : j + 1]
                j+=1
            floats.append(float(trim))
        else:
            floats.append(0.0)
    return floats

v = stringToFloat(v)
b = stringToFloat(b)
m = stringToFloat(m)
r = stringToFloat(r)
r0 = stringToFloat(r0)
r1 = stringToFloat(r1)
m0 = stringToFloat(m0)
m1 = stringToFloat(m1)
a = stringToFloat(a)
e = stringToFloat(e)
```

```python
def extractOutput(rawString):
    retList = []
    for j in range(1, len(rawString) - 1):
        if(rawString[j:j+3] == ' L0'):
            stop = getLenNum(rawString[j+4:])
            retList.append(rawString[j + 4 : j + 4 + stop])
        if(rawString[j:j+3] == '/L0'):
            stop = getLenNum(rawString[j+4:])
            retList.append(rawString[j + 4 : j + 4 + stop])
        if(rawString[j:j+3] == 'taL' and rawString[j+3:j+4] != '/'):
            stop = getLenNum(rawString[j+4:])
            retList.append(rawString[j + 4 : j + 4 + stop])
        if(rawString[j:j+2] == 'E0' and rawString[j-1:j] != '/'):
            stop = getLenNum(rawString[j+3:])
            retList.append(rawString[j + 3 : j + 3 + stop])
        if(rawString[j:j+3] == '/E0'):
            stop = getLenNum(rawString[j+4:])
            retList.append(rawString[j + 4 : j + 4 + stop])
        if(rawString[j:j+3] == 'taE' and rawString[j+3:j+4] != '/'):
            stop = getLenNum(rawString[j+4:])
            retList.append(rawString[j + 4 : j + 4 + stop])
        if(rawString[j:j+4] == 'Rmin' and rawString[j+4:j+5] != '_'):
            stop = getLenNum(rawString[j+5:])
            stop2 = getLenNum(rawString[j + 5 + stop + 2:])
            retList.append(rawString[j + 5 + stop + 2 : j + 5 + stop + 2 + stop2]) 
        if(rawString[j:j+3] == 'a_f'):
            stop = getLenNum(rawString[j+4:])
            retList.append(rawString[j + 4 : j + 4 + stop])
        if(rawString[j:j+3] == 'e_f'):
            stop = getLenNum(rawString[j+4:])
            retList.append(rawString[j + 4 : j + 4 + stop])
            break
            
    for i in range(len(retList)):
        retList[i] = float(retList[i])  
    return retList     
```
